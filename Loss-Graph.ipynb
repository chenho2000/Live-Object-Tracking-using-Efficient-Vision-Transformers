```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

## ETTracker

```python
with open('./ETTracker_results/losses.pkl', 'rb') as file:
    training_losses = pickle.load(file)
```

```python
#Avg train loss for last epoch per batch
np.mean(training_losses[-1091:])
```

```python
len(training_losses)/10
```

```python
# Plot the training losses
plt.figure(figsize=(10, 5))
plt.plot(training_losses, label='Training Losses')
plt.title('Training Loss (ETTracker)')
plt.xlabel('Batches')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
loss_per_epoch = []
for i in range(10):
    x = np.mean(training_losses[i*1091:(i+1)*1091])
    loss_per_epoch.append(x)
```

```python
loss_per_epoch
```

```python
with open('./ETTracker_results/ettracker_min_loss_test_losses.pkl', 'rb') as file:
    testing_losses_min_loss = pickle.load(file)
```

```python
np.mean(testing_losses_min_loss)
```

```python
with open('./ETTracker_results/ettracker_test_losses.pkl', 'rb') as file:
    testing_losses_final_model = pickle.load(file)
```

```python
np.mean(testing_losses_final_model)
```

```python
# Plot the testing losses
plt.figure(figsize=(10, 5))
plt.plot(testing_losses_min_loss, label='Testing Losses (Min Loss Model)', color='r')
plt.plot(testing_losses_final_model, label='Testing Losses (Final Model)', color='g')
plt.title('Testing Loss (ETTracker)')
plt.xlabel('Batches')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
